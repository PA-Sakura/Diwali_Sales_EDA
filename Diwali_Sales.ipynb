```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape') 
```

```python
df.shape
```

```python
df.head(10)
```

```python
# Data cleaning
# df.info() gives the overview of entire data with datatypes and the count using which we can eliminate null data columns
df.info()
```

```python
# Drop unrelated or blank or null columns 
# Keep the column names which you want to drop in the drop command
# axis=1 means complete column 
# inplace= true is to implement the command

df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# check null values, .sum() gives the number of null in each column
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# drop null values 
# inplace = True is necessary to save the changes made to the data
df.dropna(inplace= True)
```

```python
df.shape
```

```python
# Change the data type of the columns with float to int
# Columns names are alwasys in Square brackets
df['Amount']= df['Amount'].astype('int')
```

```python
#CHecking datatype of a column Amount
df['Amount'].dtypes
```

```python
# To get all columns names
df.columns
```

```python
# Rename columns 
# Dictionary is alwasys inside (parameter={'keyvalue':'pair'})
# this will not be saved as we dint use inplace or equalsto operator.
df.rename(columns = {'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
# use describe for specific columns
df[['Age','Orders']].describe()
```

## Exploratory data analysis

### Gender

```python
df.columns
```

```python
# Count plot using seaborn
ax=sns.countplot(x='Gender',data=df)

# To get values of each gender use the following code
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To get the amount of sales made by each gernder useing groupby  
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
```

```python
Sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sns.barplot(x='Gender', y= 'Amount', data = Sales_gen)
```

#### From above we see that Females make more number of purchases when compared to males 

### Age

```python
df.columns
```

```python
# Build a count plot for the age group
# Hue is to get division based on gender 

ax=sns.countplot (data = df, x= 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age
Sales_Age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(data=Sales_Age, x='Age Group', y='Amount')
```

#### From above graph we can see that most buyers are from age 26-35 and among them most are Female

### State

```python
df.columns
```

```python
# Total number of orders for top 15 states
Sales_State= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(15)
```

```python
# Bar plot
## Size of chart for clarity
sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(data=Sales_State, x='State', y='Orders')
```

 #### Highest number of orders are coming from Uttarpradesh followed by Maharashtra and Karnataka and least number of Orders are from Rajasthan and Punjab

```python
# Total amount of sales for top 10 states
Sales_State = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
```

```python
# Barplot for amount of sales for top 10 states
sns.barplot(data=Sales_State, x='State', y='Amount')
```

#### Most amount of sales are coming from Uttarpradesh, Maharashtra and Karnataka

### Marital Status

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_State= df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
```

```python
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=Sales_State, x='Marital_Status',y='Amount',hue='Gender')
```

#### From above we see that married Women make most amount of purchases

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
# Boxplot to see the highest amount of purchases
sns.barplot(data=Sales_Occupation, x='Occupation', y='Amount')
```

#### Most buyers are from Occupations in IT Sector, Healthcare an Aviation.

### Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,8)})
ax= sns.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Product_Category=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(19,5)})
sns.barplot(data=Sales_Product_Category, x='Product_Category', y='Amount')
```

#### Most sales are made in Food category followed by clothing & apparel, Electronics & gadgets and Footwear & shoes. 

```python
Sales_Product_ID= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
```

```python
ax=sns.barplot(data=Sales_Product_ID, x= 'Product_ID', y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
```
